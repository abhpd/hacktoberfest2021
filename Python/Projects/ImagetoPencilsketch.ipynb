**STEP03 - Following operations to be performed on image so that it can convert into the pencil sketch of respective image.**







---



---


1.   Grayscale conversion.
2.   Invert the grayscale image.
3.   Blur the inverted image using GaussianBlur() function.
4.   Inverting the above blurred image.
5.   Final Pencilsketch received.






```python
#Grayscale conversion.
#grey scale image generation from real image
grey_filter = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
cv2_imshow(grey_filter)
```

```python
#Inverting the grayscale image.
invert =cv2.bitwise_not(grey_filter)
cv2_imshow(invert)
```

```python
#Bluring the inverted image using GaussianBlur() function.
blur_image=cv2.GaussianBlur(invert,(21,21),0)
cv2_imshow(blur_image)
```

```python
#Inverting the above blurred image.
invertedblur_image=cv2.bitwise_not(blur_image)
cv2_imshow(invertedblur_image)
```

```python
#Final Pencil sketch received
Pencil_sketch_filter_image=cv2.divide(grey_filter,invertedblur_image,scale=256.0)
cv2_imshow(Pencil_sketch_filter_image)
```



---
