```python
def step_gradient(points,learning_rate,m):
    m_slope=np.zeros(14)
    M=len(points)
    for i in range(M):
        x=points[i]
        y=Y[i]
        for j in range(14):
            m_slope[j] = m_slope[j]+(-2/M)* (y - (m * x).sum() )*x[j]
    
    for j in range(14):
        m[j]=m[j]-learning_rate*m_slope[j]
        
    return m   
```
